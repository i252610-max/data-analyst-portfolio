```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv(r'E:\data-analyst-portfolio\Sample-Superstore.csv', encoding='latin1')
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Order Year'] = df['Order Date'].dt.year
df['Order Month'] = df['Order Date'].dt.month

# Subplot grid — 4 charts in one figure
fig, axes = plt.subplots(2, 2, figsize=(14,10))

# Chart 1 - Sales by Region
region_sales = df.groupby('Region')['Sales'].sum()
axes[0,0].bar(region_sales.index, region_sales.values, color='steelblue')
axes[0,0].set_title('Sales by Region')

# Chart 2 - Profit by Category
cat_profit = df.groupby('Category')['Profit'].sum()
axes[0,1].barh(cat_profit.index, cat_profit.values, color=['red','orange','green'])
axes[0,1].set_title('Profit by Category')

# Chart 3 - Sales distribution
axes[1,0].hist(df['Sales'], bins=50, color='steelblue', edgecolor='white')
axes[1,0].set_title('Sales Distribution')

# Chart 4 - Discount vs Profit
axes[1,1].scatter(df['Discount'], df['Profit'], alpha=0.2, color='coral')
axes[1,1].axhline(y=0, color='red', linestyle='--')
axes[1,1].set_title('Discount vs Profit')

plt.tight_layout()
plt.savefig('matplotlib_dashboard.png')
plt.show()
```
