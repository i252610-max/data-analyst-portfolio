```python
# Open a new notebook: day10_pandas_practice.ipynb
# Practice these Pandas operations from memory on Superstore data

import pandas as pd
df = pd.read_csv(r'E:\data-analyst-portfolio\Sample-Superstore.csv', encoding='latin1')

# 1 - Rename columns
df = df.rename(columns={'Sales': 'Total_Sales', 'Profit': 'Net_Profit'})
print(df.columns)

# 2 - Create new calculated columns
df['Profit_Ratio'] = df['Net_Profit'] / df['Total_Sales']
df['High_Value'] = df['Total_Sales'] > df['Total_Sales'].mean()

# 3 - Filter and group
high_value_by_region = df[df['High_Value']].groupby('Region')['Total_Sales'].sum()
print(high_value_by_region)

# 4 - Sort and get top 10
top_products = df.groupby('Product Name')['Net_Profit'].sum().sort_values(ascending=False).head(10)
print(top_products)
```
