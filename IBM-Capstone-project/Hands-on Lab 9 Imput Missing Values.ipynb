### Step 1: Import Required Libraries


```python
import pandas as pd
```

### Step 2: Load the Dataset Into a Dataframe


#### **Read Data**
<p>
The functions below will download the dataset into your browser:
</p>


```python
# Load dataset with remote->local fallback
remote_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/UDKAZw-kz18Yj8P6icf_qw/survey-data-duplicates.csv"
local_file = 'survey-data-duplicates.csv'
try:
    df = pd.read_csv(remote_url)
    print('Loaded remote dataset')
except Exception as e:
    print(f'Could not load remote dataset ({e}). Using local file: {local_file}')
    df = pd.read_csv(local_file)

# Display the first few rows to ensure it loaded correctly
print('Dataset shape:', df.shape)
print(df.head())
```

### Step 3. Finding and Removing Duplicates
##### Task 1: Identify duplicate rows in the dataset.


```python
# Task 1: Identify duplicate rows
dup_count = df.duplicated().sum()
print('Number of exact duplicate rows:', dup_count)
if dup_count > 0:
    print('Examples of duplicate rows:')
    print(df[df.duplicated()].head())
```

##### Task 2: Remove the duplicate rows from the dataframe.



```python
# Task 2: Remove duplicate rows
initial_shape = df.shape
df = df.drop_duplicates().reset_index(drop=True)
print(f'Removed {initial_shape[0] - df.shape[0]} duplicate rows. New shape: {df.shape}')
```

### Step 4: Finding Missing Values
##### Task 3: Find the missing values for all columns.


```python
# Task 3: Find missing values for all columns
missing_counts = df.isna().sum()
print('Missing values by column (top):')
print(missing_counts[missing_counts > 0].sort_values(ascending=False).head(20))
print('Total missing values:', missing_counts.sum())
```

##### Task 4: Find out how many rows are missing in the column RemoteWork.


```python
# Task 4: Count missing rows in 'RemoteWork'
if 'RemoteWork' in df.columns:
    missing_remote = df['RemoteWork'].isna().sum()
    print('Missing values in RemoteWork:', missing_remote)
else:
    print('RemoteWork column not found')
```

### Step 5. Imputing Missing Values
##### Task 5: Find the value counts for the column RemoteWork.


```python
# Task 5: Value counts for 'RemoteWork'
if 'RemoteWork' in df.columns:
    print(df['RemoteWork'].value_counts(dropna=False))
else:
    print('RemoteWork column not found')
```

##### Task 6: Identify the most frequent (majority) value in the RemoteWork column.



```python
# Task 6: Identify most frequent value in 'RemoteWork'
if 'RemoteWork' in df.columns:
    mode_remote = df['RemoteWork'].mode(dropna=True)
    if not mode_remote.empty:
        print('Most frequent RemoteWork value:', mode_remote.iloc[0])
    else:
        print('No mode found for RemoteWork')
else:
    print('RemoteWork column not found')
```

##### Task 7: Impute (replace) all the empty rows in the column RemoteWork with the majority value.



```python
# Task 7: Impute missing RemoteWork with mode
if 'RemoteWork' in df.columns:
    df_imputed = df.copy()
    mode_remote = df_imputed['RemoteWork'].mode(dropna=True)
    if not mode_remote.empty:
        df_imputed['RemoteWork'] = df_imputed['RemoteWork'].fillna(mode_remote.iloc[0])
        print('Imputed RemoteWork with mode:', mode_remote.iloc[0])
    else:
        print('No mode to impute RemoteWork')
    print('Missing RemoteWork after imputation:', df_imputed['RemoteWork'].isna().sum())
else:
    print('RemoteWork column not available for imputation')
```

##### Task 8: Check for any compensation-related columns and describe their distribution.



```python
# Task 8: Check compensation-related columns and describe distribution
comp_cols = [c for c in df.columns if 'Comp' in c or 'Compensation' in c or 'ConvertedComp' in c or 'Salary' in c or 'CompTotal' in c]
print('Compensation-related columns found:', comp_cols)
for c in comp_cols:
    print(f'\nColumn: {c}')
    try:
        numeric = pd.to_numeric(df[c], errors='coerce')
        print(numeric.describe())
    except Exception as ex:
        print('Could not describe column', c, ex)
```
