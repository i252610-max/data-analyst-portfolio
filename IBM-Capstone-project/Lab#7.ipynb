### Step 1: Import Required Libraries


```python
import pandas as pd
```

### Step 2: Load the Dataset into a DataFrame



load the dataset using pd.read_csv()


```python
# Load dataset: try remote, fallback to local sample
remote_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/UDKAZw-kz18Yj8P6icf_qw/survey-data-duplicates.csv"
local_path = 'survey-data-duplicates.csv'
import os
try:
    df = pd.read_csv(remote_url)
    print('Loaded remote dataset')
except Exception as e:
    print(f'Could not load remote dataset ({e}). Using local file: {local_path}')
    df = pd.read_csv(local_path)

print('\nFirst 5 rows:')
print(df.head())
print('\nDataset shape:', df.shape)
```

### Step 3: Identifying Duplicate Rows


**Task 1: Identify Duplicate Rows**
  1. Count the number of duplicate rows in the dataset.
  2. Display the first few duplicate rows to understand their structure.


```python
# Task 1: Identify duplicate rows
total_dup = df.duplicated().sum()
print('Total exact duplicate rows:', total_dup)
# Show sample duplicate rows (keep=False to show all occurrences)
dups = df[df.duplicated(keep=False)]
print('\nSample duplicates:')
print(dups.head())
```

### Step 4: Removing Duplicate Rows


**Task 2: Remove Duplicates**
   1. Remove duplicate rows from the dataset using the drop_duplicates() function.
2. Verify the removal by counting the number of duplicate rows after removal .


```python
# Task 2: Remove duplicate rows (exact duplicates)
before = len(df)
df_clean = df.drop_duplicates(keep='first').reset_index(drop=True)
after = len(df_clean)
print(f'Removed {before - after} exact duplicate rows')
print('Shape after deduplication:', df_clean.shape)
```

### Step 5: Handling Missing Values


**Task 3: Identify and Handle Missing Values**
   1. Identify missing values for all columns in the dataset.
   2. Choose a column with significant missing values (e.g., EdLevel) and impute with the most frequent value.


```python
# Task 3: Identify and handle missing values
print('Missing values by column before imputation:')
print(df_clean.isna().sum())
# Impute EdLevel with most frequent value
if 'EdLevel' in df_clean.columns:
    mode = df_clean['EdLevel'].mode(dropna=True)
    if not mode.empty:
        df_clean['EdLevel'] = df_clean['EdLevel'].fillna(mode.iloc[0])
        print('\nImputed missing EdLevel with mode:', mode.iloc[0])
    else:
        print('\nNo mode found for EdLevel; skipping imputation')
print('\nMissing values by column after imputation:')
print(df_clean.isna().sum())
```

### Step 6: Normalizing Compensation Data


**Task 4: Normalize Compensation Data Using ConvertedCompYearly**
   1. Use the ConvertedCompYearly column for compensation analysis as the normalized annual compensation is already provided.
   2. Check for missing values in ConvertedCompYearly and handle them if necessary.


```python
# Task 4: Normalize compensation data using ConvertedCompYearly
if 'ConvertedCompYearly' in df_clean.columns:
    df_clean['ConvertedCompYearly'] = pd.to_numeric(df_clean['ConvertedCompYearly'], errors='coerce')
    missing_comp = df_clean['ConvertedCompYearly'].isna().sum()
    print('Missing ConvertedCompYearly before fill:', missing_comp)
    median_comp = df_clean['ConvertedCompYearly'].median()
    df_clean['ConvertedCompYearly'] = df_clean['ConvertedCompYearly'].fillna(median_comp)
    print('Filled missing ConvertedCompYearly with median:', median_comp)
    print('\nConvertedCompYearly stats:')
    print(df_clean['ConvertedCompYearly'].describe())
else:
    print('ConvertedCompYearly column not found')
```

### Step 7: Summary and Next Steps

- You handled missing values by imputing the most frequent value in a chosen column.

- You used ConvertedCompYearly for compensation normalization and handled missing values.

```python
# Final: show cleaned dataframe and summary
print('\nCleaned DataFrame head:')
print(df_clean.head())
print('\nCleaned DataFrame shape:', df_clean.shape)
print('\nNumber of missing values:')
print(df_clean.isna().sum())
```
