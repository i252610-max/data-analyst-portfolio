```python
import plotly.express as px
import pandas as pd
import plotly.io as pio


df = pd.read_csv(r'E:\data-analyst-portfolio\Sample-Superstore.csv', encoding='latin1')

fig = px.bar(
    df.groupby('Region')['Sales'].sum().reset_index(),
    x='Region', y='Sales', title='Sales by Region', color='Region'
)

pio.renderers.default = "browser"
fig.show()

fig2 = px.scatter(
    df.sample(1000), x='Sales', y='Profit',
    color='Category', size='Quantity',
    title='Sales vs Profit by Category'
)
fig2.show()
```
