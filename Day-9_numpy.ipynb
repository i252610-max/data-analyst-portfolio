```python
# Cell 1 - imports and load data
import pandas as pd
import numpy as np

df = pd.read_csv(r'E:\data-analyst-portfolio\Sample-Superstore.csv', encoding='latin1')

# Cell 2 - basic exploration
print("Shape:", df.shape)
print(df.head())

# Cell 3 - NumPy analysis
sales_array = df['Sales'].values
profit_array = df['Profit'].values

print("Mean sales:", np.mean(sales_array))
print("Median sales:", np.median(sales_array))
print("Std deviation:", np.std(sales_array))
print("Max sale:", np.max(sales_array))
print("Min sale:", np.min(sales_array))

correlation = np.corrcoef(sales_array, profit_array)
print("\nCorrelation matrix:\n", correlation)
```
