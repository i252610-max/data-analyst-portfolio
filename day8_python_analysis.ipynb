```python
import pandas as pd
import numpy as np

# Load data
df = pd.read_csv(r'E:\data-analyst-portfolio\Sample-Superstore.csv', encoding='latin1')

# Question 1: Which sub-category has highest average profit?
subcategory_profit = df.groupby('Sub-Category')['Profit'].mean().sort_values(ascending=False)
print("Average Profit by Sub-Category:\n", subcategory_profit)

# Question 2: How many orders does each segment have?
segment_orders = df['Segment'].value_counts()
print("\nOrders by Segment:\n", segment_orders)

# Question 3: What is the profit margin for each category?
df['Profit_Margin'] = (df['Profit'] / df['Sales']) * 100
category_margin = df.groupby('Category')['Profit_Margin'].mean()
print("\nProfit Margin % by Category:\n", category_margin)
```
