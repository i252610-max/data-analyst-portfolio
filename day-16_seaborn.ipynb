Chart 1 — Distribution plot with KDE:

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv(r'E:\data-analyst-portfolio\Sample-Superstore.csv', encoding='latin1')

plt.figure(figsize=(10,5))
sns.histplot(df['Sales'], bins=50, kde=True, color='steelblue')
plt.title('Sales Distribution with KDE')
plt.savefig('sales_kde.png')
plt.show()
```

Chart 2 — Boxplot:

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='Region', y='Profit', data=df, palette='Set2')
plt.title('Profit Distribution by Region')
plt.savefig('profit_boxplot.png')
plt.show()
```

Chart 3 — Scatter with regression line:

```python
plt.figure(figsize=(10,6))
sns.regplot(x='Sales', y='Profit', data=df.sample(500), 
            scatter_kws={'alpha':0.3}, line_kws={'color':'red'})
plt.title('Sales vs Profit with Regression Line')
plt.savefig('regression_plot.png')
plt.show()
```

Chart 4 — Heatmap:

```python
plt.figure(figsize=(8,6))
corr = df[['Sales','Quantity','Discount','Profit']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.savefig('heatmap.png')
plt.show()
```

Chart 5 — Countplot:

```python
plt.figure(figsize=(8,5))
sns.countplot(x='Category', hue='Region', data=df, palette='Set1')
plt.title('Orders by Category and Region')
plt.savefig('countplot.png')
plt.show()
```

Chart 6 — Violinplot:

```python
plt.figure(figsize=(10,6))
sns.violinplot(x='Category', y='Sales', data=df, palette='muted')
plt.title('Sales Distribution by Category')
plt.savefig('violinplot.png')
plt.show()
```
