Chart 1 — Bar chart: Sales by Region


```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv(r'E:\data-analyst-portfolio\Sample-Superstore.csv', encoding='latin1')

region_sales = df.groupby('Region')['Sales'].sum().sort_values(ascending=False)

plt.figure(figsize=(8,5))
plt.bar(region_sales.index, region_sales.values, color='steelblue')
plt.title('Total Sales by Region')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.tight_layout()
plt.savefig('region_sales.png')
plt.show()
```

Chart 2 — Horizontal bar: Profit by Category

```python
category_profit = df.groupby('Category')['Profit'].sum().sort_values()

plt.figure(figsize=(8,4))
plt.barh(category_profit.index, category_profit.values, color=['red','orange','green'])
plt.title('Total Profit by Category')
plt.xlabel('Total Profit')
plt.tight_layout()
plt.savefig('category_profit.png')
plt.show()
```

Chart 3 — Scatter plot: Sales vs Profit


```python
plt.figure(figsize=(8,5))
plt.scatter(df['Sales'], df['Profit'], alpha=0.3, color='steelblue')
plt.title('Sales vs Profit')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.axhline(y=0, color='red', linestyle='--')
plt.tight_layout()
plt.savefig('sales_vs_profit.png')
plt.show()
```

Chart 4 — Pie chart: Orders by Segment


```python
segment_orders = df['Segment'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(segment_orders.values, labels=segment_orders.index, autopct='%1.1f%%')
plt.title('Orders by Customer Segment')
plt.tight_layout()
plt.savefig('segment_orders.png')
plt.show()
```

Chart 5 — Histogram: Sales distribution

```python
plt.figure(figsize=(8,5))
plt.hist(df['Sales'], bins=50, color='steelblue', edgecolor='white')
plt.title('Distribution of Sales Values')
plt.xlabel('Sale Amount')
plt.ylabel('Frequency')
plt.tight_layout()
plt.savefig('sales_distribution.png')
plt.show()
```
